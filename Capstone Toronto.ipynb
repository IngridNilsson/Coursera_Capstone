```python
table = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', na_values="Not assigned")
table[0]

df=table[0]

df = df[df['Borough'].notna()]

df.reset_index(inplace = True)

del df['index']

df
```

```python
df.shape[0]
```

### Get the latitude and the longitude coordinates of each neighborhood

```python
df1 = pd.read_csv('http://cocl.us/Geospatial_data')
df1
```

```python
df2 = df.merge(df1,how='inner', on='Postal Code')
df2
```

### Creating a map and cluster boroughs 

#### Only include Boroughs where "Toronto" is a part of the name 

```python
df3 = df2[df['Borough'].str.contains('Toronto',regex=False)]
df3
```

```python
map_toronto = folium.Map(location=[43.651070, -79.347015], zoom_start=12)

markers_colors = []
for lat, lon, borough, neighbourhood in zip(df3['Latitude'], df3['Longitude'], df3['Borough'], df3['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color='green',
        fill=True,
        fill_color='green',
        fill_opacity=0.7).add_to(map_toronto)
       
map_toronto
```

##### Creating a datafram with clustered borough and name it toronto_clustering

```python
kclusters = 5

toronto_clustering = df3.drop(['Postal Code', 'Borough','Neighbourhood'],1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
df3.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df3
```

```python
map_toronto_clusters = folium.Map(location=[43.651070, -79.347015], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Neighbourhood'], df3['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_toronto_clusters)
       
map_toronto_clusters
```
